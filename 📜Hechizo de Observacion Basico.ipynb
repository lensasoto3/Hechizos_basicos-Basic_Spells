```python
import psutil
from datetime import datetime

def estado_sistema():
    print("🧙‍♂️ Estado del sistema:\n")
    
    # Hora actual
    ahora = datetime.now()
    print(f"⏰ Hora: {ahora.strftime('%Y-%m-%d %H:%M:%S')}")
    
    # CPU
    cpu = psutil.cpu_percent(interval=1)
    print(f"⚙️ CPU en uso: {cpu}%")
    
    # RAM
    ram = psutil.virtual_memory()
    print(f"🧠 RAM usada: {ram.percent}%")

    # Alerta de CPU alto
    if cpu > 80:
        print("⚠️ Alerta: Uso de CPU alto!")

if __name__ == "__main__":
    estado_sistema()
```
